### Desacar de Yahoo Finance el historico de NVDA

```python
import yfinance as yf
start_date = '2025-01-01'
end_date = '2025-08-06'

nvda = yf.Ticker("INTC")
df = nvda.history(start=start_date, end=end_date, auto_adjust=True)

df

print(df.head())
print(df.isna().sum())
print(df.shape)
```

### Calcular indicadores tcnicos

### RSI (14)

```python
def compute_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI'] = compute_rsi(df)
df['RSI'] = df['RSI'].fillna(method='bfill')  # o 'ffill'


df.head(200)
```

### MACD

```python
exp1 = df['Close'].ewm(span=12, adjust=False).mean()
exp2 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = exp1 - exp2
df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

df.head(200)
```

### Grfico de MACD

```python
# Clculo del MACD
ema12 = df['Close'].ewm(span=12, adjust=False).mean()
ema26 = df['Close'].ewm(span=26, adjust=False).mean()
macd = ema12 - ema26
signal = macd.ewm(span=9, adjust=False).mean()
histograma = macd - signal

import matplotlib.pyplot as plt

# Colores ms suaves
soft_green = '#66bb6a'
soft_red = '#ef5350'

bar_colors = [soft_green if h >= 0 else soft_red for h in histograma]

# Estilo
plt.style.use('dark_background')

fig, ax = plt.subplots(figsize=(19, 5))

# Histograma sin borde
ax.bar(df.index, histograma, color=bar_colors, width=1.5, alpha=0.7, label='Histograma', edgecolor='none')

# Lneas MACD y Signal
ax.plot(df.index, macd, color='dodgerblue', linewidth=1.5, label='MACD')
ax.plot(df.index, signal, color='orange', linewidth=1.5, label='Signal')

# Eliminar borde blanco del grfico
for spine in ax.spines.values():
    spine.set_visible(False)

# Mejorar esttica
ax.set_title('MACD (12, 26, 9)')
ax.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()

plt.show()

```

### Modelo de prediccin (regresin lineal)

```python
# Crear columnas de features
for i in range(1, 6):
    df[f'Close_lag_{i}'] = df['Close'].shift(i)

df.dropna(inplace=True)

# Features y target
X = df[[f'Close_lag_{i}' for i in range(1, 6)]]
y = df['Close']

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

# Modelo de regresin lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Prediccin de test y evaluacin
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```

### Prediccin a futuro (prximos 5 das)

```python
# ltimos 5 precios para iniciar
last_days = df['Close'].iloc[-5:].values.reshape(1, -1)

# Obtener fecha del ltimo dato
last_date = df.index[-1]

# Fechas hbiles siguientes
next_business_days = pd.bdate_range(start=last_date + BDay(1), periods=5)

# Guardar predicciones
future_preds = []

# Prediccin iterativa
for i in range(5):
    pred = model.predict(last_days)[0]
    future_preds.append(pred)
    last_days = np.append(last_days[:, 1:], [[pred]], axis=1)

# Mostrar con seales
print(" Predicciones con seales:\n")
for i in range(len(future_preds)):
    date = next_business_days[i].date()
    price = future_preds[i]

    # Definir la seal
    if i == 0:
        prev = df['Close'].iloc[-1]
    else:
        prev = future_preds[i - 1]

    if price > prev:
        signal = " Comprar"
    elif price < prev:
        signal = " Vender"
    else:
        signal = " Mantener"

    print(f"{date}  Precio estimado: ${price:.2f}  Seal: {signal}")

```

## Usar indicadores tcnicos como variables

```python
# Filtramos las columnas necesarias para el modelo
features = ['Open', 'High', 'Low', 'Close']
df = df[features]

# Escalamos solo esas columnas
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Crear secuencias (ventanas) para entrenar
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        X.append(data[i-time_step:i])
        y.append(data[i, 0:4])  # Asumiendo que queremos predecir las primeras 4 columnas
    return np.array(X), np.array(y)

time_step = 60
X, y = create_dataset(scaled_data, time_step)

# 4. Redimensionar (ya est en forma correcta: [samples, time steps, features])
print("X shape:", X.shape)  # (samples, 60, 5)
print("y shape:", y.shape)  # (samples, 4)

#  Manten solo esta seccin (la anterior estaba duplicada)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

### Crear y entrenar el modelo LSTM

```python
# 5. Dividir en entrenamiento y test
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 6. Crear el modelo LSTM multisalida
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(4))  # salida: Open, High, Low, Close

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

### Predecir los prximos 5 das

```python
# 7. Entrenar
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# 8. Guardar modelo (opcional)
# model.save("modelo_nvda_multisalida.h5")
```
