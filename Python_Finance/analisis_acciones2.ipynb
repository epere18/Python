```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

### Desacar de Yahoo Finance el historico de NVDA

```python
import yfinance as yf

start_date = '2025-01-01'
end_date = '2025-08-01'

nvda = yf.Ticker("NVDA")
df = nvda.history(start=start_date, end=end_date, auto_adjust=True)

df

print(df.head())
print(df.isna().sum())
print(df.shape)
```

### Calcular indicadores técnicos

### RSI (14)

```python
def compute_rsi(data, window=14):
    delta = data['Close'].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    
    avg_gain = gain.rolling(window=window).mean()
    avg_loss = loss.rolling(window=window).mean()
    
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

df['RSI'] = compute_rsi(df)
df.head()

# Remover zona horaria si la tiene
if df.index.tz is not None:
    df.index = df.index.tz_convert(None)

# Obtener la fecha actual
hoy = pd.Timestamp.today().normalize()  # Solo deja año-mes-día

df_filtrado = df[df.index.normalize() == "2025-07-31"]
df_filtrado


```

### MACD

```python
exp1 = df['Close'].ewm(span=12, adjust=False).mean()
exp2 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = exp1 - exp2
df['Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

df.head(200)
```

### Modelo de predicción (regresión lineal)

```python
# Crear columnas de features
for i in range(1, 6):
    df[f'Close_lag_{i}'] = df['Close'].shift(i)

df.dropna(inplace=True)

# Features y target
X = df[[f'Close_lag_{i}' for i in range(1, 6)]]
y = df['Close']

# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

# Modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Predicción de test y evaluación
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```

### Predicción a futuro (próximos 5 días)

```python
from pandas.tseries.offsets import BDay
import numpy as np
import pandas as pd

# Últimos 5 precios para iniciar
last_days = df['Close'].iloc[-5:].values.reshape(1, -1)

# Obtener fecha del último dato
last_date = df.index[-1]

# Fechas hábiles siguientes
next_business_days = pd.bdate_range(start=last_date + BDay(1), periods=5)

# Guardar predicciones
future_preds = []

# Predicción iterativa
for i in range(5):
    pred = model.predict(last_days)[0]
    future_preds.append(pred)
    last_days = np.append(last_days[:, 1:], [[pred]], axis=1)

# Mostrar con señales
print("📈 Predicciones con señales:\n")
for i in range(len(future_preds)):
    date = next_business_days[i].date()
    price = future_preds[i]

    # Definir la señal
    if i == 0:
        prev = df['Close'].iloc[-1]
    else:
        prev = future_preds[i - 1]

    if price > prev:
        signal = "📈 Comprar"
    elif price < prev:
        signal = "🔻 Vender"
    else:
        signal = "⏸ Mantener"

    print(f"{date} → Precio estimado: ${price:.2f} → Señal: {signal}")

```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout


# Filtramos las columnas necesarias para el modelo
features = ['Open', 'High', 'Low', 'Close']
df = df[features]

# Escalamos solo esas columnas
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Crear secuencias (ventanas) para entrenar
def create_dataset(data, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        X.append(data[i-time_step:i])
        y.append(data[i, 0:4])  # Asumiendo que queremos predecir las primeras 4 columnas
    return np.array(X), np.array(y)

time_step = 60
X, y = create_dataset(scaled_data, time_step)

# 4. Redimensionar (ya está en forma correcta: [samples, time steps, features])
print("X shape:", X.shape)  # (samples, 60, 5)
print("y shape:", y.shape)  # (samples, 4)

# ✅ Mantené solo esta sección (la anterior estaba duplicada)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

### Crear y entrenar el modelo LSTM

```python
# 5. Dividir en entrenamiento y test
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 6. Crear el modelo LSTM multisalida
model = Sequential()
model.add(LSTM(64, return_sequences=True, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dropout(0.2))
model.add(Dense(4))  # salida: Open, High, Low, Close

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

### Predecir los próximos 5 días

```python
# 7. Entrenar
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# 8. Guardar modelo (opcional)
# model.save("modelo_nvda_multisalida.h5")
```

```python
# 9. Predecir próximos 5 días
last_60_days = scaled_data[-60:, :]  # Tomamos las últimas 60 filas de todas las columnas
input_seq = last_60_days.reshape(1, 60, scaled_data.shape[1])  # Asegura (1, 60, 5)

predictions = []
for _ in range(10):
    pred = model.predict(input_seq, verbose=0)[0]  # pred: [open, high, low, close]
    predictions.append(pred)

    # Crear siguiente paso: copiar última fila y actualizar columnas 0:4
    next_step = input_seq[0, -1, :].copy()
    next_step[0:4] = pred  # columnas predichas (Open, High, Low, Close)
    input_seq = np.append(input_seq[:, 1:, :], next_step.reshape(1, 1, -1), axis=1)

# Convertir predicciones a array
predictions = np.array(predictions)  # shape: (5, 4)

# Crear matriz de 5 filas y todas las columnas (para desescalar bien)
last_values = scaled_data[-1, :]  # Última fila completa, shape: (5,)
predicted_full = np.tile(last_values, (10, 1))  # Repetimos para cada predicción

# Reemplazamos solo las columnas predichas (0:4)
predicted_full[:, 0:4] = predictions

# Invertir la escala
predicted_real = scaler.inverse_transform(predicted_full)[:, 0:4]  # Tomamos Open, High, Low, Close

# Fechas de los próximos 5 días hábiles
predicted_dates = pd.date_range(df.index[-1] + pd.Timedelta(days=1), periods=10, freq='B')

# Mostrar resultados
for date, row in zip(predicted_dates, predicted_real):
    print(f"{date.strftime('%Y-%m-%d')}: " +
          f"Open: ${row[0]:.2f}, High: ${row[1]:.2f}, Low: ${row[2]:.2f}, Close: ${row[3]:.2f}")
```
